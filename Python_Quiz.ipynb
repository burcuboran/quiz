```python
import json
import datetime
from time import time
import pymysql.cursors

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='admin',
                             db='quiz',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)


firstName=input("Enter your name: ")
lastName=input("Enter your surname: ")
totalScore=0

try: 
    with connection.cursor() as cursor:
        # Read a single record
        sql = "SELECT * FROM `User` WHERE firstName=%s AND lastName=%s"
        cursor.execute(sql,(firstName,lastName))
        result = cursor.fetchone()
        if(result != None):
            totalScore = result["totalScore"]
            print("Your current score is: ", totalScore)
        else:
            with connection.cursor() as cursor:
                sql = "INSERT INTO `User` (`firstName`, `lastName`,`totalScore`) VALUES (%s, %s, %s)"
                cursor.execute(sql, (firstName, lastName, totalScore))     
            connection.commit()

 # timenow = datetime.datetime.now()
    
    start_time = time()

    # Opening JSON file 
    f = open('./quiz.json') 

    # returns JSON object as a dictionary 
    quiz = json.load(f) 

   

    # Iterating through the json list 
    for i in quiz['questions']: 
        print(i['question'])
        print("answers:")
        for key, value in i['answers'].items():
            print(key,":",value)
        ans=input("Enter your answer: ")   
        if  ans == i['correctAnswers']:
            print("Bravo correct answer is ",i['correctAnswers'])
            totalScore+=1     
        #elif ans not in key:
            #raise("please neter a letter between a and d")
        else:
            print("Sorry, wrong answer")
            
       

    end_time = time() - start_time
    minute = int(end_time / 60)
    seconds = int(end_time % 60)
    print("Your timing is: ", minute, ':', seconds)
    print("Your final score is: ", totalScore)

    with connection.cursor() as cursor:
        sql = "UPDATE  User SET totalScore = %s WHERE firstName=%s AND lastName=%s"
        cursor.execute(sql, (0,firstName, lastName))     
    connection.commit()

finally:
    connection.close()
    
# Closing file 
f.close()
```
